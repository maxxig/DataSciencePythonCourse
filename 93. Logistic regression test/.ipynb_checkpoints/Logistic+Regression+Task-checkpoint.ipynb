### 5. Use seaborn to create a histogram of the Age column.

```python
sns.histplot(data=data_set,x =data_set['Age'], bins=30)
```

### 6. Use seaborn to create a jointplot showing Area Income versus Age.

```python
sns.jointplot(data=data_set, x=data_set['Age'], y=data_set['Area Income'], s=60)
```

### 7. Use seaborn to create a jointplot  showing the kde distributions of Daily Time Spent on Site versus Age.

```python
sns.jointplot(data = data_set
              , x = data_set['Age']
              , y = data_set['Daily Time Spent on Site']
              , kind = 'kde'
              , cmap='hot_r'
              , n_levels=60
              , fill=True
              , marginal_kws={'color':'red'})
```

### 8. Use seaborn to create a jointplot of Daily Time Spent on Site versus Daily Internet Usage.

```python
sns.jointplot(data=data_set
              , x=data_set['Daily Time Spent on Site'] 
              , y=data_set['Daily Internet Usage']
              , color ='r'
              , s=50
             )
```
