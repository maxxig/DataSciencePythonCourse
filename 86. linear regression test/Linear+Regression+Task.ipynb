### 16. Now let's split the data into training and testing sets. Set a variable X equal to the numerical features of the customers and a variable y equal to the "Yearly Amount Spent" column.

```python
X=df_original[['Avg. Session Length', 'Time on App','Time on Website', 'Length of Membership']]
```

```python
Y=df_original['Yearly Amount Spent']
```

### 17. Use model_selection.train_test_split from sklearn to split the data into training and testing sets. Set test_size=0.4 and random_state=101

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.4, random_state=101)
```

### 18. Now its time to train the model on training data. Import LinearRegression from sklearn.linear_model 

```python
from sklearn.linear_model import LinearRegression
```

### 19. Create an instance of a LinearRegression() model and assign to a variable

```python
lrm = LinearRegression()
```

### 20. Train/fit the model on the training data

```python
lrm.fit(X_train,y_train)
```

### 21. Print out the coefficients of the model

```python
lrm.coef_
```

### 22. Now that we have fit our model, let's evaluate its performance by predicting off the test values. Use predict() to predict off the X_test set of the data

```python
predictions = lrm.predict(X_test)
#predictions
```

### 23. Create a scatterplot of the real test values vs the predicted values

```python
plt.scatter(y_test,predictions)
```

### 24. Calculate the Mean Absolute Error, Mean Squared Error, and the Root Mean Squared Error

```python
from sklearn import metrics
message = 'MAE:' + str(metrics.mean_absolute_error(y_test,predictions)) + ' MSE'+str(metrics.mean_squared_error(y_test,predictions))+' RMSE ' + str(np.sqrt(metrics.mean_squared_error(y_test,predictions)))
message
```

### 25. Let's quickly explore the residuals to make sure everything was okay with our data. Plot a histogram of the residuals and make sure it looks normally distributed. Use either seaborn distplot, or just plt.hist()

```python
#plt.hist((y_test - predictions))
sns.displot((y_test - predictions), kde=True, bins=50)
```
