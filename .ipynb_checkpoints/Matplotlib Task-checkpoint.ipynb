# Matplotlib Task

## Create plots using the following data

```python
import numpy as np
a = np.arange(0,100)
b = a - 5
c = a ** 3 - b ** 2
```

### 1. Import matplotlib.pyplot and set %matplotlib inline

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

### 2. a. Create a figure object
### b. Add axes to the figure object passing [0, 0, 1, 1] list as a parameter
### c. Create a plot with a as x axis data and b as y axis data
### d. Add labels to the plot like on the plot below

```python
figure = plt.figure()
axes = figure.add_axes([0,0,1,1])
axes.plot(a,b)
axes.set_title('its my lable')
```

### 3. Create a figure object and add two axes - axes1 and axes2. Pass the following lists as parameters - [0, 0, 1, 1] to axes1 and [0.2, 0.5, 0.2, 0.3] to axes2

```python
figure1 = plt.figure()
axes1 = figure1.add_axes([0,0,1,1])
axes2 = figure1.add_axes([0.2,0.5,0.2,0.3])
```

### 4. Create plots with a as x axis data and b as y axis data on axes1 and with b as x axis data and c as y axis data on axes2 for the previos figure object and show the plot by calling figure variable

```python
figure1 = plt.figure()
axes1 = figure1.add_axes([0,0,1,1])
axes2 = figure1.add_axes([0.2,0.5,0.2,0.3])
axes1.plot(a,b)
axes2.plot(b,c)
```

```python
figure2 = plt.figure()
axes21 = figure2.add_axes([0,0,1,1])
axes22 = figure2.add_axes([0.2,0.5,0.3,0.4])
axes21.set_xlim([20,60])
axes21.set_ylim([10,70])
axes22.set_xlim([20,60])
axes22.set_ylim([10000,70000])
axes22.set_xlabel('a array')
axes22.set_ylabel('c array')
axes21.plot(a,b)
axes21.set_xlabel('a array')
axes21.set_ylabel('b array')
axes22.plot(a,c)
axes21.set_title('Outer plot')
axes22.set_title('Inner plot')
```
